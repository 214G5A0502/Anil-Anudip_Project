```python
import mysql.connector
import pandas as pd

# Connect to MySQL database
conn = mysql.connector.connect(host="localhost", user="root", passwd="Ma@8978120031", database="anil")

# Define SQL query
query = "SELECT * FROM ethnic_wear"

# Read data into Pandas DataFrame
df = pd.read_sql(query, conn)

# Display the DataFrame
print(df.head())

# Close the connection
conn.close()
```

```python
import matplotlib.pyplot as plt

# Histogram of Ratings
plt.hist(df['ratings'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Ratings')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram of Discount Prices
plt.hist(df['discount_price'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Discount Prices')
plt.xlabel('Discount Prices')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar chart of Top Main Categories by Count
top_main_categories = df['main_category'].value_counts().head(10)
top_main_categories.plot(kind='bar', color='orange')
plt.title('Top Main Categories by Count')
plt.xlabel('Main Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar chart of Average Ratings by Main Category
avg_ratings_by_main_category = df.groupby('main_category')['ratings'].mean().sort_values(ascending=False)
avg_ratings_by_main_category.plot(kind='bar', color='purple')
plt.title('Average Ratings by Main Category')
plt.xlabel('Main Category')
plt.ylabel('Average Ratings')
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot of Ratings vs No. of Ratings
plt.scatter(df['ratings'], df['no_of_ratings'], color='red', alpha=0.5)
plt.title('Correlation between Ratings and No. of Ratings')
plt.xlabel('Ratings')
plt.ylabel('No. of Ratings')
plt.show()
```
